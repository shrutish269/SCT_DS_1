```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the gender_submission file
df = pd.read_csv('/content/gender_submission.csv')
df.head()
```

```python
print("Shape:", df.shape)
df.info()
df.describe()
```

```python
print("Missing Values:\n", df.isnull().sum())
```

```python
# Countplot for survival predictions
sns.countplot(data=df, x='Survived')
plt.title("Survival Prediction Distribution")
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.ylabel("Number of Passengers")
plt.show()

# Pie chart
survived_counts = df['Survived'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(survived_counts, labels=['Did Not Survive', 'Survived'], autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
plt.title("Survival Prediction Pie Chart")
plt.show()
```
