```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/global_student_migration.csv')
df.head()
```

```python
# Step 3: Group by destination_country and count number of students
country_counts = df['destination_country'].value_counts().reset_index()
country_counts.columns = ['destination_country', 'student_count']

# Step 4: Plot the bar chart
plt.figure(figsize=(12,6))
sns.barplot(x='destination_country', y='student_count', data=country_counts, palette='viridis')
plt.title('Number of International Students by Destination Country')
plt.xlabel('Destination Country')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Drop rows with missing salary data
df = df.dropna(subset=['starting_salary_usd'])

# Plot histogram for starting salary
plt.figure(figsize=(10,6))
sns.histplot(df['starting_salary_usd'], bins=15, kde=True, color='teal')
plt.title('Distribution of Starting Salaries (USD)')
plt.xlabel('Starting Salary (USD)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
